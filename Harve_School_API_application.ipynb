```python
import datetime # Import the datetime module

# Simular que vamos buscar dados de vários dias (ou de múltiplas APIs)
cotacoes = []

for i in range(3):  # Fazer 3 requisições
    response = requests.get('https://api.hgbrasil.com/finance?key=3b9b39d6')
    data = response.json()

    # Adicionar timestamp da requisição à extração
    dolar = data['results']['currencies']['USD']
    dolar['timestamp'] = datetime.datetime.now().isoformat() # Record current time
    dolar['moeda'] = 'USD'

    cotacoes.append(dolar)

df_cotacoes = pd.DataFrame(cotacoes)
print(df_cotacoes)
```

```python
# Simular um arquivo XML com dados de clientes
xml_content = """<?xml version="1.0" encoding="UTF-8"?>
<clientes>
    <cliente>
        <id>001</id>
        <nome>João Silva</nome>
        <pais>Brasil</pais>
        <moeda>USD</moeda>
        <valor_contrato>10000</valor_contrato>
    </cliente>
    <cliente>
        <id>002</id>
        <nome>Maria Santos</nome>
        <pais>Brasil</pais>
        <moeda>USD</moeda>
        <valor_contrato>25000</valor_contrato>
    </cliente>
    <cliente>
        <id>003</id>
        <nome>Pedro Costa</nome>
        <pais>Brasil</pais>
        <moeda>EUR</moeda>
        <valor_contrato>15000</valor_contrato>
    </cliente>
</clientes>
"""

# Salvar para simular arquivo XML
with open('clientes.xml', 'w', encoding='utf-8') as f:
    f.write(xml_content)
```

```python
import xml.etree.ElementTree as ET

# Ler o XML
tree = ET.parse('clientes.xml')
root = tree.getroot()

# Extrair dados para lista
clientes = []
for cliente in root.findall('cliente'):
    clientes.append({
        'id': cliente.find('id').text,
        'nome': cliente.find('nome').text,
        'pais': cliente.find('pais').text,
        'moeda': cliente.find('moeda').text,
        'valor_contrato': float(cliente.find('valor_contrato').text)
    })

df_clientes = pd.DataFrame(clientes)
print("✅ Clientes carregados do XML:")
print(df_clientes)
```

```python
# Fazer requisição à API
response = requests.get('https://api.hgbrasil.com/finance?key=3b9b39d6')
data = response.json()

# Extrair moedas
currencies = data['results']['currencies']

print("\n✅ Cotações obtidas da API:")
for moeda, valores in currencies.items():
    if moeda in ['USD', 'EUR', 'GBP']:
        print(f"{moeda}: Compra={valores['buy']}, Venda={valores['sell']}")
```

```python
# Preparar DataFrame com cotações
cotacoes_lista = []
for moeda, valores in currencies.items():
    # Skip the 'source' entry as it does not contain currency data like 'buy', 'sell'
    if moeda == 'source':
        continue
    cotacoes_lista.append({
        'moeda': moeda,
        'taxa_compra': valores['buy'],
        'taxa_venda': valores['sell'],
        'variacao': valores['variation']
    })

df_cotacoes = pd.DataFrame(cotacoes_lista)

# Fazer merge com os clientes
df_resultado = df_clientes.merge(df_cotacoes, on='moeda', how='left')

print("\n✅ Dados enriquecidos (XML + API):")
print(df_resultado)
```

```python
# Calcular o valor do contrato em reais (convertendo de USD/EUR)
def converter_para_reais(row):
    if row['moeda'] == 'USD':
        # Se for dólar, usar a taxa de compra do dólar
        return row['valor_contrato'] * row['taxa_compra']
    elif row['moeda'] == 'EUR':
        # Se for euro, buscar a taxa do euro
        if 'EUR' in currencies:
            return row['valor_contrato'] * currencies['EUR']['buy']
    return row['valor_contrato']

df_resultado['valor_em_reais'] = df_resultado.apply(converter_para_reais, axis=1)

# Calcular margem (contrato em reais vs valor original)
df_resultado['margem'] = df_resultado['valor_em_reais'] - df_resultado['valor_contrato']

print("\n✅ Com conversão para reais:")
print(df_resultado[['id', 'nome', 'valor_contrato', 'moeda', 'taxa_compra', 'valor_em_reais']])
```

```python
# Salvar como CSV
df_resultado.to_csv('clientes_enriquecidos.csv', index=False, encoding='utf-8')
print("✅ Arquivo salvo: clientes_enriquecidos.csv")

# Salvar como JSON
df_resultado.to_json('clientes_enriquecidos.json', orient='records', indent=2)
print("✅ Arquivo salvo: clientes_enriquecidos.json")

# Exibir resumo
print("\n📊 RESUMO DO ETL:")
print(f"Total de clientes: {len(df_resultado)}")
print(f"Valor total de contratos (original): {df_resultado['valor_contrato'].sum():,.2f}")
print(f"Valor total em reais: R$ {df_resultado['valor_em_reais'].sum():,.2f}")
```

A nossa empresa importa peças de tecnologia. Nós temos uma lista de produtos onde alguns são comprados em Dólar (USD) e outros em Euro (EUR).

A Missão de vocês:

Criar o DataFrame inicial com os dados fornecidos.

Fazer uma requisição na API de finanças da HG Brasil para pegar a cotação de compra (buy) do USD e do EUR hoje.

Criar uma nova coluna chamada Preco_BRL aplicando a lógica: Se a moeda for USD, multiplica pelo valor do Dólar. Se for EUR, multiplica pelo valor do Euro.

Exportar o resultado final para um arquivo chamado estoque_atualizado.csv

```python
#DESAFIOSS!!
import pandas as pd

dados_estoque = {
    "Produto": ["Placa Mãe", "Processador", "Memória RAM", "Monitor", "Teclado"],
    "Moeda": ["USD", "USD", "EUR", "EUR", "USD"],
    "Preco_Original": [150.00, 250.00, 80.00, 200.00, 45.00]
}
df_estoque = pd.DataFrame(dados_estoque)
```

```python
import requests
import pandas as pd

# 1. Puxar os dados da API
url = "https://api.hgbrasil.com/finance"
resposta = requests.get(url).json()

# 2. Navegar no JSON para pegar as cotações
dolar = resposta['results']['currencies']['USD']['buy']
euro = resposta['results']['currencies']['EUR']['buy']

# 3. Lógica matemática condicional no Pandas
# Cria uma lista vazia e preenche com os valores calculados
valores_reais = []
for index, linha in df_estoque.iterrows():
    if linha['Moeda'] == 'USD':
        valores_reais.append(linha['Preco_Original'] * dolar)
    elif linha['Moeda'] == 'EUR':
        valores_reais.append(linha['Preco_Original'] * euro)

# 4. Adicionar a coluna e exportar
df_estoque['Preco_BRL'] = valores_reais
df_estoque['Preco_BRL'] = df_estoque['Preco_BRL'].round(2) # Arredondar para 2 casas

df_estoque.to_csv("estoque_atualizado.csv", index=False)
print("✅ Planilha estoque_atualizado.csv gerada com sucesso!")
print(df_estoque)
```

A Missão de vocês:

Puxar os dados da API da HG Brasil e isolar apenas o bloco de stocks (bolsas de valores).

Transformar esse dicionário confuso de bolsas em um DataFrame do Pandas.

Criar uma regra de negócio (lógica condicional): Criar uma coluna chamada Status. Se a variação (variation) for maior que 0, escreva 'Alta'. Se for menor que 0, escreva 'Queda'.

Desafio Matemático: Criar uma coluna chamada Lucro_R$1000. Se eu tivesse investido R$ 1000 nessas bolsas ontem, quanto eu teria hoje baseado nessa variação? (Fórmula: 1000 + (1000 * (variation / 100)))

Salvar o relatório como relatorio_bolsas.csv

```python
import requests
import pandas as pd

# 1. Puxar dados da API
url = "https://api.hgbrasil.com/finance"
resposta = requests.get(url).json()

# 2. Isolar o dicionário de ações
bolsas = resposta['results']['stocks']

# 3. Transformar o Dicionário em Lista de Dicionários para o Pandas entender bem
lista_bolsas = []
for sigla, dados in bolsas.items():
    lista_bolsas.append({
        "Bolsa": dados['name'],
        "Local": dados['location'],
        "Pontos": dados['points'],
        "Variacao_Perc": dados['variation']
    })

df_bolsas = pd.DataFrame(lista_bolsas)

# 4. Lógica de Status (Alta ou Baixa) usando apply e lambda (ou função simples)
def define_status(variacao):
    if variacao > 0:
        return 'Alta'
    elif variacao < 0:
        return 'Queda'
    else:
        return 'Estável'

df_bolsas['Status'] = df_bolsas['Variacao_Perc'].apply(define_status)

# 5. Desafio Matemático: Cálculo do investimento de R$ 1000
df_bolsas['Lucro_R$1000'] = 1000 + (1000 * (df_bolsas['Variacao_Perc'] / 100))
df_bolsas['Lucro_R$1000'] = df_bolsas['Lucro_R$1000'].round(2)

# 6. Exportar
df_bolsas.to_csv("relatorio_bolsas.csv", index=False, encoding='utf-8')
print("✅ Relatório de bolsas gerado com sucesso!")
display(df_bolsas)
```
