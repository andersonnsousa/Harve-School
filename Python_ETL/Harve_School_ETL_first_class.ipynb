```python
import pandas as pd

# ============================================================
# EXTRAÇÃO
# ============================================================
df_titanic = pd.read_csv('https://www.harve.com.br/praticas/titanic-pt-BR.csv')

print("=" * 55)
print("EXTRAÇÃO")
print("=" * 55)
print(f"Registros carregados: {len(df_titanic)}")
print(f"\nValores nulos por coluna:\n{df_titanic.isnull().sum()}")
```

```python
# ============================================================
# TRANSFORMAÇÃO — Desafio 1
# ============================================================
print("\n" + "=" * 55)
print("TRANSFORMAÇÃO — Desafio 1: Local de embarque")
print("=" * 55)

# Registros sem local de embarque
sem_embarque = df_titanic[df_titanic['Local de embarque'].isnull()]
print(f"\nRegistros sem local de embarque ({len(sem_embarque)}):")
print(sem_embarque[['Id do passageiro', 'Nome', 'Local de embarque']])

# Excluir registros sem local de embarque
df_titanic = df_titanic.dropna(subset=['Local de embarque'])
print(f"\nRegistros após exclusão: {len(df_titanic)}")
print(f"Nulos em 'Local de embarque': {df_titanic['Local de embarque'].isnull().sum()}")
```

```python
# ============================================================
# TRANSFORMAÇÃO — Desafio 1 (continuação)
# ============================================================
print("\n" + "=" * 55)
print("TRANSFORMAÇÃO — Desafio 1: Idade nula → média")
print("=" * 55)

idades_nulas_antes = df_titanic['Idade'].isnull().sum()
media_idade = df_titanic['Idade'].mean()
print(f"Idades nulas antes: {idades_nulas_antes}")
print(f"Média de idade calculada: {media_idade:.2f}")

df['Idade'] = df_titanic['Idade'].fillna(media_idade)

print(f"Idades nulas após preenchimento: {df_titanic['Idade'].isnull().sum()}")
```

```python
# ============================================================
# TRANSFORMAÇÃO — Desafio 2: Categorização por idade
# ============================================================
print("\n" + "=" * 55)
print("TRANSFORMAÇÃO — Desafio 2: Categoria de idade")
print("=" * 55)

def categorizar_idade(idade):
    if idade <= 10:
        return 'Criança'
    elif idade <= 20:
        return 'Joven'
    elif idade <= 60:
        return 'Adulto'
    else:
        return 'Idoso'

df_titanic['Categoria de Idade'] = df_titanic['Idade'].apply(categorizar_idade)

print("\nContagem por categoria (value_counts):")
print(df_titanic['Categoria de Idade'].value_counts())
```

```python
# ============================================================
# CARGA
# ============================================================
print("\n" + "=" * 55)
print("CARGA")
print("=" * 55)

output_path = 'titanic_transformado.csv'
df.to_csv(output_path, index=False, encoding='utf-8-sig')
print(f"Arquivo salvo em: {output_path}")
print(f"Total de registros exportados: {len(df_titanic)}")
print(f"\nPrimeiras linhas do resultado:")
print(df_titanic[['Id do passageiro', 'Nome', 'Idade', 'Local de embarque', 'Categoria de Idade']].head(5))
```

```python
import pandas as pd

# ============================================================
# EXTRAÇÃO
# ============================================================
df_netflix = pd.read_csv('http://www.harve.com.br/praticas/NetflixViewingHistory.csv')

print("=" * 55)
print("EXTRAÇÃO")
print("=" * 55)
print(f"Registros carregados: {len(df_netflix)}")
print(f"\nPrimeiras linhas:")
print(df_netflix.head())

# ============================================================
# TRANSFORMAÇÃO
# ============================================================

# --- Converter coluna Date para datetime ---
df_netflix['Date'] = pd.to_datetime(df_netflix['Date'], dayfirst=True)

# ============================================================
# Desafio 1: Quantos são filmes e quantos são séries?
# Lógica: títulos com ':' no nome são séries, sem ':' são filmes
# ============================================================
print("\n" + "=" * 55)
print("DESAFIO 1: Filmes vs Séries")
print("=" * 55)

df_netflix['Tipo'] = df_netflix['Title'].apply(
    lambda titulo: 'Série' if ':' in titulo else 'Filme'
)

contagem_tipo = df_netflix['Tipo'].value_counts()
print(f"\nContagem por tipo:")
print(contagem_tipo)

total = len(df_netflix)
for tipo, qtd in contagem_tipo.items():
    print(f"  {tipo}: {qtd} ({qtd/total*100:.1f}%)")

# ============================================================
# Desafio 2: Dia da semana com mais e menos visualizações
# 0: Segunda  1: Terça  2: Quarta  3: Quinta
# 4: Sexta    5: Sábado  6: Domingo
# ============================================================
print("\n" + "=" * 55)
print("DESAFIO 2: Dia da semana")
print("=" * 55)

dias_semana = {
    0: 'Segunda',
    1: 'Terça',
    2: 'Quarta',
    3: 'Quinta',
    4: 'Sexta',
    5: 'Sábado',
    6: 'Domingo'
}

df_netflix['Dia da Semana Num'] = df_netflix['Date'].dt.dayofweek
df_netflix['Dia da Semana']     = df_netflix['Dia da Semana Num'].map(dias_semana)

views_por_dia = df_netflix['Dia da Semana'].value_counts()
views_por_dia_pct = df_netflix['Dia da Semana'].value_counts(normalize=True) * 100

dia_mais  = views_por_dia.idxmax()
dia_menos = views_por_dia.idxmin()

print(f"\nVisualizações por dia da semana:")
for dia in dias_semana.values():
    qtd = views_por_dia.get(dia, 0)
    pct = views_por_dia_pct.get(dia, 0)
    print(f"  {dia:<10}: {qtd:>3} visualizações ({pct:.1f}%)")

print(f"\n✅ Dia com MAIS  visualizações: {dia_mais}  ({views_por_dia[dia_mais]} | {views_por_dia_pct[dia_mais]:.1f}%)")
print(f"✅ Dia com MENOS visualizações: {dia_menos} ({views_por_dia[dia_menos]} | {views_por_dia_pct[dia_menos]:.1f}%)")

# ============================================================
# CARGA
# ============================================================
print("\n" + "=" * 55)
print("CARGA")
print("=" * 55)

output_path = 'netflix_transformado.csv'
df_netflix.to_csv(output_path, index=False, encoding='utf-8-sig')
print(f"Arquivo salvo em: {output_path}")
print(f"Total de registros exportados: {len(df_netflix)}")
print(f"\nColunas finais: {list(df_netflix.columns)}")
```
